```python
import numpy as np 
import pandas as pd  
import os 
import os
import numpy as np
import pandas as pd
from toto.evaluation.fev.evaluate import DATASETS
from toto.evaluation.fev.utils import (aggregate_per_dataset_model, 
                                       compute_overall_model_agg, 
                                       compute_rank_summary, 
                                       get_datasets_metadata, 
                                       make_latex_table_from_dict)
```

```python
RESULTS_PATH = "./results"
METRICS = ["abs_error", "mean_wQuantileLoss", "MASE"]   # lower is better

# models to include in the aggregation and comparison
MODELS_TO_INCLUDE = ['toto', 'toto_finetuning', 'toto_finetuning_exogenous']
```

```python
# Aggregation (gmean) of results per dataset and model
dataset_model_agg = aggregate_per_dataset_model(
    RESULTS_PATH,
    DATASETS,
    MODELS_TO_INCLUDE,
    METRICS
)

dataset_model_agg
```

```python
def show_aggregated_results(overall_model_agg: pd.DataFrame, rank_summary: pd.DataFrame, metrics: list[str]):
    print("\nOverall aggregates per model (gmean over datasets, per-metric filtering):")
    columns = ["model"] + metrics
    display(overall_model_agg[columns])

    print("\nRank summary (lower is better):")
    display(rank_summary)

# get overall aggregated results on all datasets
overall_model_agg = compute_overall_model_agg(dataset_model_agg, METRICS)
rank_summary = compute_rank_summary(dataset_model_agg, METRICS)
show_aggregated_results(overall_model_agg, rank_summary, METRICS)
```

```python
# collect fev datasets metadata for latex table formatting
datasets_metadata = get_datasets_metadata("./tasks.yaml", DATASETS)
```

```python

# ----------------------------
# Writing results to latex
# ----------------------------
METRICS = ["mean_wQuantileLoss", "MASE"]
METRIC_DISPLAY = {"mean_wQuantileLoss": "WQL", "MASE": "MASE"}

MODELS = ["ZS", "FT", "FT & EV"]

# display name -> internal model key (must match dataset_model_agg["model"])
MODEL_KEY = {
    "ZS": "toto",  # change if needed
    "FT": "toto_finetuning",
    "FT & EV": "toto_finetuning_exogenous",
}

# your dict keys -> latex headers (and order)
META_MAP = [
    ("num_rows", "# rows"),
    ("series_length", "length"),
    ("num_evs", "# EVs"),
    ("horizon", "horizon"),
]

# Example:
tex = make_latex_table_from_dict(
    dataset_model_agg=dataset_model_agg,
    datasets_metadata=datasets_metadata,
    model_key=MODEL_KEY,
    metric_display=METRIC_DISPLAY,
    meta_map=META_MAP,
    out_tex_path="results_table.tex",
    section_title=f"Aggregated results on {len(datasets_metadata)} datasets",
    decimals=3,
)
print("Wrote results_table.tex")
```
